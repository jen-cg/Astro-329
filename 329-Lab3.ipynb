```python
def t(du, sky, area_app , S_N):
    '''Time required to reach a certain signal to noise ratio in Victoria'''
    #area_ann = pi*( 40**2 - 30**2 )
    
    a = du *(3/2)
    b = sky*area_app*(3/2) #sky*area_ann*(3/2)
    c = ( 1.4628271535796697 * 3)**2 * (area_app)
    R = S_N
    
    term1 = (a + b)*(R**2)
    term2 = 4*(a**2)*(c)*(R**2)
    term3 = 2*(a**2)
    
    return [ (term1 + sqrt( term1**2 + term2 ))/(term3) , (term1 - sqrt( term1**2 + term2 ))/(term3) ]

def t_Vic(M , S_N):
    return t( 10**( m*M + b ) , msky_dupix , area_16 , S_N  )
```

```python
M = 20
S_N = [10,20,50,100]
print('Victoria')
for i in S_N:
    time = t_Vic(M,i)
    print('S/N {} : Time in min {} ,  {}'.format(i, time[0]/60 , time[1]/60 ))
```

```python
def t_dark(du, area_app , S_N):
    '''Time required to reach a certain signal to noise ratio in a dark site'''
    #area_ann = pi*( 40**2 - 30**2 )
    
    a = du *(3/2)
    b =  (22.5)*(0.435**2)*(3/2)*area_app#(area_ann)
    c = ( 1.4628271535796697 * 3)**2 * (area_app)
    R = S_N
    
    term1 = (a + b)*(R**2)
    term2 = 4*(a**2)*(c)*(R**2)
    term3 = 2*(a**2)
    
    return [ (term1 + sqrt( term1**2 + term2 ))/(term3) , (term1 - sqrt( term1**2 + term2 ))/(term3) ]

def t_dark_wrap(M , S_N):
    return t_dark( 10**( m*M + b ) , area_16 , S_N  )
```

```python
M = 20
S_N = [10,20,50,100]
print('Dark site')
for i in S_N:
    time = t_dark_wrap(M,i)
    print('S/N {} : Time in min {} ,  {}'.format(i, time[0]/60 , time[1]/60 ))
```

```python
M = 20
S_N = [10,20,50,100]
print('Dark site')
for i in S_N:
    time_Vic = t_Vic(M,i)
    time_dark = t_dark_wrap(M,i)
    diff = time_Vic[0]/60 - time_dark[0]/60
    print('S/N {} : Time in min {}'.format(i, diff))
```

# Faintest star that can be observed

```python
def a(tt,S_N,area_app,sky):
    
    b = (sky)*(3/2)*area_app
    c = ( 1.4628271535796697 * 3)**2 * (area_app)
    R = S_N
    
    term1 = R**2 * tt
    term2 = 4*(tt**2)*( tt*b* (R**2) + c* (R**2) )
    term3= 2*(tt**2)
    
    return [ (term1 + sqrt( term1**2 + term2 ))/(term3) , (term1 - sqrt( term1**2 + term2 ))/(term3) ]

def awrap_Vic(tt,S_N):
    return a(tt,S_N,area_16,msky_dupix)

def awrap_dark(tt,S_N):
    return a(tt,S_N,area_16,(22.5)*(0.435**2))
```
